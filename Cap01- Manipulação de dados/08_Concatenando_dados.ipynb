## Seção 8 - Concatenando dados
___

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

### Concatenando dados

#### pd.concat([objeto_da_esquerda, objeto_da_direita])

♔┈♔┈♔┈( O QUE ISSO FAZ )┈♔┈♔┈♔

O método `concat()` junta dados de duas fontes diferentes em uma única tabela. O comportamento padrão é juntar as fontes considerando os índices das colunas.

```python
pd.concat([df, df2])
```

#### pd.concat([objeto_da_esquerda, objeto_da_direita], axis=1)

♔┈♔┈♔┈( O QUE ISSO FAZ )┈♔┈♔┈♔

O método `concat()` junta dados de duas fontes diferentes em uma única tabela. Passando o argumento `axis=1` para o método, o Python entende que deve juntar essas fontes usando o índice das linhas de ambas fontes

```python
pd.concat([df, df2], axis=1)
```
