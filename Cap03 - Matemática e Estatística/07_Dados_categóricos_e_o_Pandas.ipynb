```python
import numpy as np
import pandas as pd


caminho = 'C:/Users/Jessica/Desktop/roteiros/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'qtd_marchas', 'carb']
carros.index = carros.nomes
```

```python
carros_cat = carros[['cyl','vs','am','qtd_marchas','carb']]
carros_cat.head()
```

```python
marchas_grupos = carros_cat.groupby('qtd_marchas')
marchas_grupos.describe().T
```

### Transformando variáveis em dados categóricos

#### pd.Series(variável_x, dtype)

♔┈♔┈♔┈( O QUE ISSO FAZ )┈♔┈♔┈♔

Para criar uma Serie de dados categóricos usamos o `pd.Series()` passando um array ou uma série que contém o dado que queremos transformar e o argumento `dtype="categorical"`.
Isso avisa ao Python para criar uma Series que contém dados categóricos. Ao atribuirmos isso à `carros['grupo']` criamos uma nova coluna em `carros` contendo esse novo dado.

```python
carros['grupo'] = pd.Series(carros['qtd_marchas'], dtype="category")
```

```python
carros['grupo'].dtype
```

```python
carros['grupo'].head()
```

```python
carros['grupo'].value_counts()
```

### Descrevendo dados categóricos com crosstabs

#### pd.crosstab(variável_y, variável_x)

♔┈♔┈♔┈( O QUE ISSO FAZ )┈♔┈♔┈♔

Para montar um cruzamento entre váriaveis usamos o médodo do Pandas `pd.crosstab()` passando as variáveis que queremos ver no nosso resultado.

```python
pd.crosstab(carros['am'], carros['qtd_marchas'])
```
